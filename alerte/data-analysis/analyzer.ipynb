# Analyzer

In this notebook, we analyze our farms data in France to create questions for the project

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
farms = pd.read_csv("../data-mining/farms.csv")
rubrics = pd.read_csv("../data-mining/rubrics.csv")
stocks = pd.read_csv("../data-mining/stocks.csv")
```

```python
farms.info()
```

```python
group_df = farms[['name', 'type', 'mise_en_demeure']].groupby(['type', 'mise_en_demeure']).count()

group_df.unstack().plot(kind='bar', stacked=True)
plt.legend(["NON", "OUI"], title="Mise en demeure");
plt.xticks(rotation=45)
plt.ylabel("Nombre")
plt.title("Fermes en France")
plt.show()
```
